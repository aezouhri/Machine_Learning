# Title: Homework 4 Problem 3

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Common Parameters

```python
theta_0 = 1  # True parameter value
lambda_max = max(16, 100)
lambda_values = np.linspace(0, lambda_max, 500)
N_values = [2, 10, 1000]
sigma = 3  # Given variance of the noise for problems 3a to 3c
lambda_optimal = sigma ** 2 / theta_0 ** 2  # Optimal lambda for the lowest MSE
```

# Part a

```python
# Plot Bias for each N and mark the optimal lambda
for N in N_values:
    bias = - (lambda_values * theta_0) / (N + lambda_values)
    plt.plot(lambda_values, bias, label=f'N={N}')
    # Calculate the bias at the optimal lambda
    bias_optimal = - (lambda_optimal * theta_0) / (N + lambda_optimal)
    plt.scatter(lambda_optimal, bias_optimal, color='red', zorder=5, label='Optimal Lambda')  # Mark the optimal lambda

plt.xlabel('Lambda')
plt.ylabel('Bias of Ridge Estimate')
plt.title('Bias of Ridge Regression Estimate as a Function of Lambda')
plt.legend()
plt.grid(True)
plt.show()
```

# Part b

```python
# Plot Variance for each N and mark the optimal lambda
for N in N_values:
    variance = (sigma ** 2 * N) / (N + lambda_values) ** 2
    plt.plot(lambda_values, variance, label=f'N={N}')
    # Calculate the variance at the optimal lambda
    variance_optimal = (sigma ** 2 * N) / (N + lambda_optimal) ** 2
    plt.scatter(lambda_optimal, variance_optimal, color='red', zorder=5,
                label='Optimal Lambda')  # Mark the optimal lambda

plt.xlabel('Lambda')
plt.ylabel('Variance of Ridge Estimate')
plt.title('Variance of Ridge Regression Estimate as a Function of Lambda')
plt.legend()
plt.grid(True)
plt.show()
```

# Part c

```python
# Plot MSE for each N and mark the optimal lambda
for N in N_values:
    mse = (sigma ** 2 + lambda_values ** 2 * theta_0 ** 2) / (N + lambda_values) ** 2
    plt.plot(lambda_values, mse, label=f'N={N}')
    # Calculate the MSE at the optimal lambda
    mse_optimal = (sigma ** 2 + lambda_optimal ** 2 * theta_0 ** 2) / (N + lambda_optimal) ** 2
    plt.scatter(lambda_optimal, mse_optimal, color='red', zorder=5, label='Optimal Lambda')  # Mark the optimal lambda

plt.xlabel('Lambda')
plt.ylabel('MSE of Ridge Estimate')
plt.title('MSE of Ridge Regression Estimate as a Function of Lambda')
plt.legend()
plt.grid(True)
plt.show()
```

# Part f

```python
# Parameters
N = 10  # Given number of observations
theta_0 = 1  # True parameter value

# Plot Bias^2, Variance, and MSE for each sigma value
for sigma in [1, 4, 10]:
    # Calculate the bias^2, variance, and MSE
    bias_squared = ((-(lambda_values * theta_0) / (N + lambda_values)) ** 2)
    variance = (sigma ** 2 * N) / ((N + lambda_values) ** 2)
    mse = (sigma ** 2 + lambda_values ** 2 * theta_0 ** 2) / ((N + lambda_values) ** 2)

    # Calculate the optimal lambda for the lowest MSE
    lambda_optimal = sigma ** 2 / theta_0 ** 2
    mse_optimal = (sigma ** 2 + lambda_optimal ** 2 * theta_0 ** 2) / (N + lambda_optimal) ** 2

    # Create plots
    plt.figure(figsize=(10, 4))

    # Plot Bias^2, Variance, and MSE
    plt.plot(lambda_values, bias_squared, label='Bias^2')
    plt.plot(lambda_values, variance, label='Variance')
    plt.plot(lambda_values, mse, label='MSE')

    # Set the plot limits to better visualize the curves
    plt.xlim(0, lambda_max)
    plt.ylim(0, max(np.max(bias_squared), np.max(variance), np.max(mse)) * 1.1)  # 10% above the max value for clarity

    # Set plot titles and labels
    plt.xlabel('Lambda')
    plt.ylabel('Estimate')
    plt.title(f'Bias^2, Variance, and MSE for N={N}, sigma={sigma}')
    plt.legend()
    plt.grid(True)
    plt.show()
```
